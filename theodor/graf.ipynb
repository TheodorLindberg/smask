```python
import pandas as pd
import numpy as np
df = pd.read_csv("training_data.csv")
y = df["increase_stock"].map({
    "low_bike_demand": 0,
    "high_bike_demand": 1
})
X = df.drop(columns=["increase_stock"])

df["month"] = df["month"] -1
```

```python

per_day = X["day_of_week"]
unique = np.unique(per_day)

total = np.zeros(len(unique))
count = np.zeros(len(unique))

for i in range(len(per_day)):
    total[per_day[i]] += 1
    count[per_day[i]] += y[i]


ratio = count/total

import matplotlib.pyplot as plt

plt.bar(np.unique(per_day), ratio * 100)
```

```python

per_day = X["hour_of_day"]
unique = np.unique(per_day)

total = np.zeros(len(unique))
count = np.zeros(len(unique))

for i in range(len(per_day)):
    total[per_day[i]] += 1
    count[per_day[i]] += y[i]


ratio = count/total

import matplotlib.pyplot as plt

plt.bar(np.unique(per_day), ratio * 100)
```

```python

per_day = X["month"]
unique = np.unique(per_day)

total = np.zeros(len(unique))
count = np.zeros(len(unique) )

for i in range(len(per_day)):
    total[per_day[i]] += 1
    count[per_day[i]] += y[i]


ratio = count/total

import matplotlib.pyplot as plt

plt.bar(np.unique(per_day), ratio * 100)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

weather_features = ['temp', 'dew', 'humidity', 'precip', 'snow', 'snowdepth', 'windspeed', 'cloudcover', 'visibility']

for feature in weather_features:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='increase_stock', y=feature, data=df)
    plt.title(f'{feature} vs Bike Demand')
    plt.show()
```

```python
import matplotlib.pyplot as plt

weather_features = ['temp', 'dew', 'humidity', 'precip', 'snow', 'snowdepth', 'windspeed', 'cloudcover', 'visibility']

for feature in weather_features:
    # Separate the data by category

    increase = df[df['increase_stock'] == "high_bike_demand"][feature]
    no_increase = df[df['increase_stock'] == "low_bike_demand"][feature]
    
    # Create a boxplot
    plt.figure(figsize=(6, 4))
    plt.boxplot([no_increase, increase], tick_labels=['No Increase', 'Increase'])
    
    plt.title(f'{feature} vs Bike Demand')
    plt.ylabel(feature)
    plt.xlabel('increase_stock')
    plt.show()

```

```python
import matplotlib.pyplot as plt

weather_features = ['hour_of_day', 'day_of_week', 'month', 'holiday', 'weekday']

for feature in weather_features:
    # Separate the data by category

    increase = df[df['increase_stock'] == "high_bike_demand"][feature]
    no_increase = df[df['increase_stock'] == "low_bike_demand"][feature]
    
    # Create a boxplot
    plt.figure(figsize=(6, 4))
    plt.boxplot([no_increase, increase], tick_labels=['No Increase', 'Increase'])
    
    plt.title(f'{feature} vs Bike Demand')
    plt.ylabel(feature)
    plt.xlabel('increase_stock')
    plt.show()

```

```python

weather_features = ['hour_of_day', 'day_of_week', 'month', 'holiday', 'weekday']
df = pd.read_csv("training_data.csv")

# Map labels to integers if needed
df['increase_stock_label'] = df['increase_stock'].map({
    "low_bike_demand": 0,
    "high_bike_demand": 1
})

corr = df[weather_features + ['increase_stock_label']].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation between weather and bike demand")
plt.show()
```
