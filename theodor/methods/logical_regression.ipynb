```python
import numpy as np

def knn_predict(X_train, y_train, x_new, k=5):
    # Compute distances
    distances = np.abs(X_train - x_new)

    # Get indices of k nearest neighbors
    knn_idx = np.argsort(distances)[:k]

    # Majority vote
    return np.round(np.mean(y_train[knn_idx]))
```
